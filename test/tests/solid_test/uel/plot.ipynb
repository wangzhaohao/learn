```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
moose = pd.read_csv('reference_out.csv')
uel = pd.read_csv('test_out.csv')
```

```python
plt.plot(uel['time'], uel['delta_l'], label = 'uel')
plt.plot(moose['time'], moose['delta_l'], label = 'MOOSE')
plt.plot(moose['time'], moose['time']**2/10, label = '$t^2$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('time')
plt.ylabel('elongation')
plt.legend()
```
